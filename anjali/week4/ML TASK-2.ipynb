### Split the data into features (X) and labels (y)

```python
x=A[['Energy','Resistance','Current','Voltage','Power','uipvalue','year','month','day']] 
y=A['prediction']
```

### Split the data into training and testing sets

```python
from sklearn.linear_model import LogisticRegression                #importing the model from sklearn
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

### Create and train/fit a logistic regression model

```python
lr= LogisticRegression()
lr.fit(x_train, y_train)
```

### Predict labels on the test set

```python
lrpred = lr.predict(x_test)
lrpred
```

### Evaluate accuracy (optional)

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, lrpred)
print(f"Accuracy: {accuracy}")
```
