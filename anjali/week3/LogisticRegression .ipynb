GridSearchCV for hyperparameter tuning on a logistic regression model.
This is a good approach to find the optimal hyperparameter values for your model.

#### Import required libraries for hyperparameter tuning from sklearn

```python
from sklearn.model_selection import GridSearchCV
```

#### Define the hyperparameter grid

```python
c_space = np.logspace(-5, 8, 15)     #c_space is a range of values for the hyperparameter C.
param_grid = {'C': c_space}    #param_grid is a dictionary specifying the hyperparameters to search and their respective ranges.
```

#### Create a logistic regression model

```python
lr = LogisticRegression()
```

#### Create a GridSearchCV object

```python
logreg_cv = GridSearchCV(lr, param_grid, cv=5) 
#GridSearchCV performs a cross-validated grid search over the hyperparameter grid, trying all possible combinations.
```

#### Fit the model with hyperparameter tuning on the training data

```python
logreg_cv.fit(x_train, y_train)
```

#### Print the best hyperparameters and the corresponding score

```python
#logreg_cv.best_params_ gives you the best hyperparameters found during the search
print("Tuned Logistic Regression Parameters: {}".format(logreg_cv.best_params_))
print("Best score is {}".format(logreg_cv.best_score_))
```
