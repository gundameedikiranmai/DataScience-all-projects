```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
t=pd.read_csv("C:\\Users\\intel\\Downloads\\Manpower_SF_Nov12.csv")
```

```python
#t=pd.to_datetime(t['attend_date'],format='%d-%m-%Y')
```

```python
t.drop(columns=["plan_cd","dept","division", "direct_indirect",
                 "employee_class","workforce_category_descr","attend_reason_code",
                 "plant_line_running"],inplace=True)
```

```python
t.rename(columns={'plant_code':'classes','absentee':'cost','total_attend_hours':'hours',
                  'jobtitle':'job','attend_type_code':'code'},inplace=True)
```

```python
#t['attend_date'] = pd.to_datetime(t['attend_date'])
#start_date = pd.to_datetime('2023-10-01')
#end_date = pd.to_datetime('2023-11-30') 
#t1 = t[(t['attend_date'] >= start_date) & (t['attend_date'] <= end_date)]
#t1
```

```python
start_date = ('2023-10-01')
end_date =('2023-11-30')
t1=t[(t['attend_date'] >= start_date) & (t['attend_date'] <= end_date)]
t1
```

```python
A=t1.groupby(['attend_date']).agg('sum').reset_index()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x="attend_date",y="cost",data=A,marker="o",ms=3,mfc="red",mec="black",lw=1)
f1={'family':'Cambria','size':'18','color':'red'}
plt.xlabel("Attend_date",size=16,color="red",fontdict=f1)
plt.ylabel("Cost",size=16,color="red",fontdict=f1)
plt.title("LINE PLOT(attend_date Vs Cost)",fontdict=f1)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x="attend_date",y="hours",data=A)
plt.xlabel("Attend_date",size=15,color="red",fontdict=f1)
plt.ylabel("Hours",size=15,color="red",fontdict=f1)
plt.title("LINE PLOT(attend_date vs hours)",color="red",size=15,fontdict=f1)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="attend_date",y="cost",data=A,palette="Dark2",edgecolor="k")
plt.xlabel("ATTEND_DATE",size=14,color="blue",fontdict=f1)
plt.ylabel("COST",size=14,color="blue",fontdict=f1)
plt.title("BAR PLOT(ATTEND_DATE Vs COST)",color="blue",size=18,fontdict=f1)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="attend_date",y="hours",data=A,palette="viridis",edgecolor="k")
plt.xlabel("Attend_date",size=14,color="red")
plt.ylabel("Hours",size=14,color="red")
plt.title("BAR PLOT(Attend_date vs hours)",color="red",size=18)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x="cost",y="hours",data=A,color="red",marker=".",s=100)
plt.xlabel("Cost",size=13,color="green",fontdict=f1)
plt.ylabel("Hours",size=13,color="green",fontdict=f1)
plt.title("SCATTER PLOT(Cost vs Hours)",color="green",size=15,fontdict=f1)
```

```python
sns.boxplot(x='job',y='hours',hue='classes',data=t1)   #hue=classes
plt.xlabel("Job",size=14,color="red")
plt.ylabel("Cost",size=14,color="red")
plt.title("BOX PLOT(Job vs Cost)",color="red",size=12)
```

```python
sns.boxplot(x='job',y='hours',hue='code',data=t1)
plt.xlabel("Job",size=14,color="red")
plt.ylabel("Hours",size=14,color="red")
plt.title("BOX PLOT(Job vs Hours)",color="red",size=12)
```
