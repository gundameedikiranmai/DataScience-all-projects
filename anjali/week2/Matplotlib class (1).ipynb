### SIMPLE/LINE PLOT 
By importing the matplotlib module,defines x and y values for a plotPython, plots the data using the plot()function and  it helps to display the plot by using the show() function.The plot() creates a line plot by connecting the points defined       by x and y values.

```python
import matplotlib.pyplot as plt

#From matplotlib import pyplot as plt
import numpy as np
```

```python
X =np.arange(5,30,5)
Y= [i**2 for i in X]
plt.plot(X,Y)
print(X,Y)
plt.title("Line plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis(x-squres)")
plt.show()
```

### Markers (representation,size and colour)
>Marker :representation of arguments

>Marker parameters: 1. Marker style(O-circle,S-square,D-diamand,P-pentagon,h-hexagon,"."-point,etc...)
                
                 2.Color:The color should be represents in a.single character(r-red,b-blue,k-black,c-cyan,etc...)
                                                           b.Hexadecimal(#RGB)
                                              >> Marker edge color(mec),marker face color(mfc)
                 3.Size(ms="give required size")
>Syntax: plt.plot(x,y, marker="o", ms="10", mfc="r", mec="k")

```python
X =np.arange(5,30,5)
Y= [i**2 for i in X]
plt.plot(X,Y,marker="o",ms="5",mfc="r",ls="None")   #for marker representation -marker=("style of marker"),marker size=ms,colour=mfc
print(X,Y,)
plt.title("Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis(x-squres)")
plt.show()
```

### Lines(line style,line width,line colour) 
Line parameter :

             1.Line style(ls):
                   ls="solid"
                   ls="dotted" :for dottled lines(.........)
                   ls="dashed" :for dashed lines(----------)
                   ls="dashdot":for dashdot lines(-.-.-)
                   ls="None" :line will not be appear
             2.line width(lw): lw="required color"
             3.line color(c): c="give any color here"

```python
X =np.arange(5,30,5)
Y= X*2
plt.plot(X,Y,marker="o",ms="5",mfc="r",ls="dashed",lw="2",c="k")   #for linestyle=ls,line width=lw,line colour=c
plt.title("Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis(x-squres)")
plt.show()
```

### Subplot function
It is used to represent multiple plots in a single figure.

Subplot function will take 3 arguments i.e rows,coloums,index of plot(subplot(R,C,I)).

Syntax: sublot(R,C,I)  -for subplots

Systax: suptitle("super title name")  -For major title 


```python
x =[0,10,20,30,40]
y1 =[300,500,200,400,100]
y2 =[100,400,300,500,200]
y3 =[200,500,100,300,400]
plt.suptitle("MULTIPLE PLOTS")

plt.subplot(1,3,1)
plt.plot(x,y1,color="g",marker="*",mfc="r",ms="8")
plt.title("PLOT-1")
plt.xlabel("Low value")
plt.ylabel("High value")

plt.subplot(1,3,2)
plt.plot(x,y2,marker="o",ms="4",mfc="b",c="m",lw="1")
plt.title("PLOT-2")
plt.xlabel("Low value")
plt.ylabel("High value")
plt.subplot(1,3,3)

plt.title("PLOT-3")
plt.plot(x,y3,color="b",marker="x",ms="5",mfc="r")
plt.xlabel("Low value")
plt.ylabel("High value")

plt.subplots_adjust(left=0.1,right=1.5,bottom=0.4,top=0.9,wspace=0.4,hspace=0.5) #for space betweenthe subplots
plt.show()
```

## Scatterplot
It is used to represent each and every point on the plot without using line(y using dots).

```python
import random
X=[1,2,3,4,5,6]
Y1=[10,20,30,40,50,60]
random.shuffle(X)
print(X,Y1)
plt.scatter(X,Y1,color="red",marker="D",s=40,linewidth=2,alpha=0.4,label="2021")

X=[2,4,6,8,10,12]
Y2=[20,40,60,80,100,120]
random.shuffle(X)
print(X,Y2)
plt.scatter(X,Y2,color="black",marker=".",s=50,linewidth=3,alpha=0.7,label="2022")
plt.legend()
plt.title("Satter plot")
plt.show()
```

 different cmaps: cmap="viridis", cmap="Accent", cmap="Rainbow"

## Pie chart
  >Pie chart is a circular,statistial chart which is divided into a different slices based on different divisions and 
    this each slice is called widge.
       function =plt.pie()
       
    parameters:
         1.labels
         2.startangle
         3.explode: it is used to explode the particular part from the chart(ex: explode=[0.2,0,0,0]).
         4.shadow : shadow= True/False
         5.color
         6.autopct: it is used to give percentage value for that particular part(ex:autopct="%2.1f%%")
         
  >plt.figure(figsize =(1,2)) : this syntax is used to increasing the figure size and it will give above the plt.pie() function.       

```python
category=["Teachers","Girls","Boys","Workers"]
data =[20,50,30,10]
plt.figure(figsize=(7,10))
plt.pie(data,labels=category,startangle=0,explode=[0,0,0.1,0.1],shadow=True,colors=["red","blue","pink","green"],autopct="%2.1f%%")
        
plt.legend(title="Info",loc="upper right")
plt.title("PIE CHART")
plt.show()
```

## Histogram graph:
 >It is an accurate representation of distribution of numerical data.In this we have to divide that numerical data into different intervals.
 
 function: plt.hist()
 
     Parameters:
      1.Histtype :histtype="step",histtype ="bar",histtype="stepfilled"
      2.rwidth   :by using this rwidth we have to divide individuals bars 
      3.facecolor
>For step and stepfilled histtype we no need to wrote the rwidth

```python
values =[98,55,43,14,50,45,32,60,88]
values_intervals=[0,40,80,100]
plt.title("")
plt.hist(values,values_intervals,histtype="bar",rwidth=0.7,facecolor="b")
plt.xticks([0,40,80,100])   
plt.xlabel("Intervals")
plt.ylabel("No.of values")
plt.title("HISTOGRAM CHART")
plt.show()
```
