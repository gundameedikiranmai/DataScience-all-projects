```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
A=pd.read_csv("C:\\Users\\intel\\Downloads\\WL_task2_ProdData_2023_versionD.csv")
A
```

### Capture the range of each feature(power,energy,voltage,resistane)
       For capturing the range(ex:Mean,median,minimum,maximum,) mostly we have to use boxplot

### Power

```python
sns.boxplot(x="id",y="Power",data=A,palette="CMRmap")
f1={'family':'Cambria','size':'14','color':'red'}
plt.xlabel("Id",size=14,color="red",fontdict=f1)
plt.ylabel("Power",size=14,color="red",fontdict=f1)
plt.title("BOX PLOT",fontdict=f1)
```

### Energy

```python
sns.boxplot(x="id",y="Energy",data=A,palette="seismic")
plt.xlabel("Id",size=14,color="red",fontdict=f1)
plt.ylabel("Energy",size=14,color="red",fontdict=f1)
plt.title("BOX PLOT",fontdict=f1)
```

### Voltage

```python
sns.boxplot(x="id",y="Voltage",data=A,palette="cool")
plt.xlabel("Id",size=14,color="red",fontdict=f1)
plt.ylabel("Voltage",size=14,color="red",fontdict=f1)
plt.title("BOX PLOT",fontdict=f1)
```

### Resistance

```python
sns.boxplot(x="id",y="Resistance",data=A,palette="winter")
plt.xlabel("Id",size=14,color="red",fontdict=f1)
plt.ylabel("Resistance",size=14,color="red",fontdict=f1)
plt.title("BOX PLOT",fontdict=f1)
```

### Capture the count of predictions

```python
a=sns.countplot(x="prediction",data=A,palette="cool",edgecolor="black")
for p in a.patches:
    a.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
          ha='center', va='baseline')
```

### Capture the data of feature score throught out the dates(serie)

### By using Seaborn

```python
A1=A['Date'] = pd.to_datetime(A['Date'])
A1=A_sorted = A.sort_values(by='Date')
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(x="Date",y="score",data=A1)
plt.xlabel("Date",size=18,color="red",fontdict=f1)
plt.ylabel("Score",size=18,color="red",fontdict=f1)
plt.title("LINE PLOT",size=20,fontdict=f1)
plt.show()
```
