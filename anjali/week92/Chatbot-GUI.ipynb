```python
import tkinter as tk
from tkinter import messagebox

#import langchain
import openai
from langchain.llms import OpenAI

warnings.filterwarnings("ignore")

#import os

class ChatApp:
    def __init__(self, master):
        self.master = master
        master.title("Chat with AI")

        # Set OpenAI API key
        os.environ["OPENAI_API_KEY"] = "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"
        #os.environ["OPEN_API_KEY"] = "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"

        # Create OpenAI instance
        self.llm = OpenAI(openai_api_key=os.environ["OPENAI_API_KEY"], temperature=0.6)

        self.chat_display = tk.Text(master, width=60, height=20, bg="#f0f0f0", bd=0, fg="#333333", font=("Arial", 10), wrap="word")
        self.chat_display.grid(row=0, column=0, columnspan=2, padx=10, pady=10)

        self.entry = tk.Entry(master, width=60, font=("Arial", 10))
        self.entry.grid(row=1, column=0, padx=10, pady=10)

        self.ask_button = tk.Button(master, text="Ask", command=self.ask_question, width=10, font=("Arial", 10))
        self.ask_button.grid(row=1, column=1, padx=10, pady=10)

        self.quit_button = tk.Button(master, text="Quit", command=master.quit, width=10, font=("Arial", 10))
        self.quit_button.grid(row=2, column=1, padx=10, pady=10)

        # Configure grid weights to allow resizing
        master.columnconfigure(0, weight=1)
        master.rowconfigure(0, weight=1)

    def ask_question(self):
        question = self.entry.get()
        if question.strip() != "":
            self.chat_display.insert(tk.END, "You: " + question + "\n", "user")
            response = self.llm.invoke(input=question)
            self.chat_display.insert(tk.END, "Chatbot: " + response + "\n", "bot")
        else:
            messagebox.showerror("Error", "Please enter a question.")
        self.entry.delete(0, tk.END)  # Clear the entry field

def main():
    root = tk.Tk()
    app = ChatApp(root)
    # Apply styles for user and bot messages
    app.chat_display.tag_configure("user", justify="right")
    app.chat_display.tag_configure("bot", justify="left")
    root.mainloop()

if __name__ == "__main__":
    main()
```
