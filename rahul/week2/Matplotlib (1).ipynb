# Markers

markers are styles,colour and sizes
style=(o)circle,(s)square,(D)diamond,(.)point,(p)pentagon,(*)star,(+)plus,(|)vline(_)hline[marker="*"]

```python
x=np.arange(0,20,2)
y=x**2
print(x,y)
plt.plot(x,y,marker="*",ms="20",mec="y",mfc="k")
plt.title("sample of markers")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=np.arange(0,5)
y=x**2
y1=x+3
y2=[11,40,20,15,24]
y3=[10,2,4,5,6]


plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("map")

plt.subplot(2,2,2)
plt.plot(x,y1)
plt.title("pop")


plt.subplot(2,2,3)
plt.plot(x,y2)
plt.title("land")

plt.subplot(2,2,4)
plt.plot(x,y3)
plt.title("earth")
plt.subplots_adjust(left=0.1,right=0.9,bottom=0.1,top=0.9,wspace=0.4,hspace=0.4)
plt.suptitle("sample of subplots",size=15)
plt.show()

```

```python
x=np.arange(0,5,1)              #WITHOUT THE PARMETERS IN LEGEND
y=x**2
y1=x**3
print(x,y,y1)
plt.plot(x,y,label="square")
plt.plot(x,y1,label="cubes")
plt.xlabel("x-axis",size=15,color="b")
plt.ylabel("y-axis",size=15,color="b")
plt.legend()
plt.title("sample legend plot",size=15,color="g")
          

plt.show()
```

```python
#WITH PARMETERS IN LEGEND
x=np.arange(1,10)
y=x*2
y1=x++2
print(x,y,y1)

plt.plot(x,y,label="square")
plt.plot(x,y1,label="cube")
plt.legend(loc="upper left",framealpha=0.9,facecolor="w",edgecolor="k",shadow=True,fancybox=True)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("legend plot",size=20)



plt.show()
```

### stacked bar chart

```python
years=["2017","2018","2019","2020"]
placements=[150,70,50,40]

cse=[70,20,40,60]
ece=[20,40,60,70]
it=[10,70,50,60]
eee=[40,60,70,80]

it_starts=[cse[i]+ece[i] for i in range(len(cse))]
eee_starts=[it_starts[i]+it[i] for i in range(len(it))]

plt.bar(years,cse,color="k",width=0.5,label="CSE")
plt.bar(years,ece,bottom=cse,color="b",width=0.5,label="ECE")
plt.bar(years,it,bottom=it_starts,color="r",width=0.5,label="IT")
plt.bar(years,eee,bottom=eee_starts,color="g",width=0.5,label="EEE")


plt.xlabel("years",size=10)
plt.ylabel("placemensts",size=10)
plt.title("placmensts in yearwise",size=10,color="b")
plt.ylim(0,300)

plt.legend(loc=2)
plt.suptitle("stacked bar chart",size=20,color="g")

plt.show()


```

### GROUP BAR CHART

```python
years=["2017","2018","2019","2020"]

ap=[20,40,30,50]
ts=[40,39,50,40]
mp=[40,80,70,20]
tm=[20,60,70,39]
w=0.2
plt.figure(figsize=(10,5))

ap_1=np.arange(len(years))
ts_1=[i+w for i in ap_1]
mp_1=[i+w for i in ts_1]
tm_1=[i+w for i in mp_1]


plt.bar(ap_1,ap,width=w,label="AP")
plt.bar(ts_1,ts,width=w,label="TS")
plt.bar(mp_1,mp,width=w,label="MP")
plt.bar(tm_1,tm,width=w,label="TM")


plt.xticks(ap_1+w+w/2,years)
plt.xlabel("Y E A R S")
plt.ylabel("pop%")
plt.ylim(0,100)
plt.title("Pop percentages in states in yearwise")
plt.legend(shadow=True,edgecolor="k")



plt.show()
```

### pie chart

```python
wheather=["cool","rain","heavyrain","hot"]
tem=[20,50,42,70]
plt.figure(figsize=(8,8))

plt.pie(tem,labels=wheather,startangle=0,shadow=True,explode=[0.2,0,0,0],colors=["g","b","r","y"],autopct="%2.1f%%")
plt.legend(title="WHEATHER",edgecolor="k",loc=2)


plt.show()
```
