# Markers
Representation of points in graphs/charts
markers are styles,colour and sizes
style=(o)circle,(s)square,(D)diamond,(.)point,(p)pentagon,(*)star,(+)plus,(|)vline(_)hline[marker="*"]
colour we can indicate in single charater(r,b,g,c,k,m,y,p) or mec=markers edage colour[mec] and markers face colour[mfc] and mutli colour#__ [color='']
size=ms"_"

```python
x=np.arange(0,20,2)
y=x**2
print(x,y)
plt.plot(x,y,marker="*",ms="20",mec="y",mfc="k")
plt.title("sample of markers")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=np.arange(0,20,2)
y=(2,3,4,7,6,2,8,6,1,4)
print(x,y)
plt.plot(x,y,marker="p",ms="15",color="c",mec="k",mfc="g")
plt.title("sample of markers")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=np.arange(0,30,3)
y=x**2
y1=x*3
print(x,y)
plt.plot(x,y,x,y1,ls="dashdot",lw=2,c="k")
plt.title("line")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=np.arange(0,15,3)
y=x**2
print(x,y)
x1=[1,20,12,14,17,6]
y1=[10,20,30,40,50,60]
plt.plot(x,y,marker=".",ms="17",mec="r",mfc="y",ls="dashed",lw=2)
#mutli lines plt.plot(x,y,x1,y1,marker=".",ms="17",mec="r",mfc="y",ls="dashed",lw=2)
plt.plot(x1,y1,ls="dashdot",lw=5,color="k")
plt.title("lines")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Font properties on title&label 

```python
x=np.arange(0,20,4)
y=[1,4,7,8,5]
print(x,y)
h1={"family":"arial","size":15,"color":"k"}
h2={"family":"cambria","size":20,"color":"g"}
plt.plot(x,y)
plt.title("line plot",h1)
plt.xlabel("x-axis",h2)
plt.ylabel("y-axis",h1)
plt.show()
```

```python
x=np.arange(0,64,8)
y=x**2
print(x,y)
plt.plot(x,y,marker="s",color='k',mec="g",mfc="y",ms=20)
plt.title('line chart')
l1={"family":"cambria",'size':20,'color':"k"}
plt.xlabel("x-axis",l1)
plt.ylabel("y-axis",l1)
plt.show()
```

### Grid function

```python
x=[1,2,3,6,7,9,8,4]
y=[10,20,50,40,28,35,45,24]
plt.plot(x,y)
plt.grid(ls="dotted",lw=2,axis="both")
plt.xlabel("x-axis",size=15,family="cambria")
plt.ylabel("y-axis",size=15,family="cambria")
plt.title("GRID FUNCTION PLOT",color="r")
plt.show()
```

```python
x=np.arange(0,10)
y=x**2
y1=x+1
print(x,y,x,y)
plt.plot(x,y,x,y1,color="g")
plt.grid(axis="x",ls="dashed",lw=2)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

### Subplots
In fig we create a subplots 
in subplots we can create the multi plots by using the subplots
in subplots
rows
coloumns
index
plt.subplot(row,column,index)

```python
x=np.arange(0,10)
y=x*2
y1=[10,20,30,10,50,60,70,10,40,15]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("pop")

plt.subplot(2,2,2)
plt.plot(x,y1)
plt.title("map")
plt.suptitle("subplots chart",size=20,color="b")

plt.show()
```

```python
x=np.arange(0,5)
y=x**2
y1=x+3
y2=[11,40,20,15,24]
y3=[10,2,4,5,6]


plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("map")

plt.subplot(2,2,2)
plt.plot(x,y1)
plt.title("pop")


plt.subplot(2,2,3)
plt.plot(x,y2)
plt.title("land")

plt.subplot(2,2,4)
plt.plot(x,y3)
plt.title("earth")
plt.subplots_adjust(left=0.1,right=0.9,bottom=0.1,top=0.9,wspace=0.4,hspace=0.4)
plt.suptitle("sample of subplots",size=15)
plt.show()

```
