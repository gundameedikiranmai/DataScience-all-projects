### matrix

```python
df=pd.read_excel(r"C:\Users\RAHUL\Desktop\kaggledata\mart.xlsx")
```

```python
df
```

```python
df1=df.pivot_table(index="Outlet_Year",columns="Outlet_Size",values="Sales")
```

```python
df1
```

```python
sns.set_style("white")
sns.heatmap(df1,annot=True,
           fmt='.1f',
           annot_kws=dict(size=15,weight="bold"),
           linewidths=1,
           linecolor="g")
sns.despine()
plt.show()
```

```python
df1.corr()
```

```python
sns.heatmap(df1.corr(),
           vmin=-1,
           vmax=1,
           center=0,
           annot=True,
           fmt=".1f",
           annot_kws=dict(size=15,weight="bold"),
           linewidth=0.5,
           linecolor="k")

plt.show()
```

```python
sns.clustermap(df1)
plt.show()
```

```python
sns.clustermap(df1,
             # col_cluster=False,
              row_cluster=False,
              annot=True,
              fmt=".0f",standard_scale=1)

plt.show()
```
