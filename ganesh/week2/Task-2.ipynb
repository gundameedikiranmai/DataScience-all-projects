```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv(r"C:\Users\ganig\OneDrive\Desktop\Dataset\ml-2.csv")
```

```python
df=pd.DataFrame(df)
```

```python
df.head()
```

## Visualizing the range of each feature(power, energy, voltage, resistance) in Train and Test data separately

```python
sns.boxplot(y="Voltage",x="id",data=df,hue="id",palette="viridis",saturation=1)
sns.set(style="darkgrid", rc={"grid.color": "green"})
plt.title("BOX PLOT",size=25,color="red",family="cambria")
plt.show()
```

```python
sns.boxplot(y="Power",x="id",data=df,hue="id",palette="Accent",saturation=1)
plt.title("BOX PLOT",size=25,color="red",family="cambria")
plt.show()
```

```python
sns.boxplot(y="Energy",x="id",data=df,hue="id",palette="rainbow",saturation=1)
plt.title("BOX PLOT",size=25,color="red",family="cambria")
plt.show()
```

```python
sns.boxplot(y="Resistance",x="id",data=df,hue="id",saturation=1)
plt.title("BOX PLOT",size=25,color="red",family="cambria")
plt.show()
```

## Visualizing the Count of predictions

```python
sns.countplot(x="prediction",data=df,edgecolor="k",hue="id",palette="Accent")
plt.title("COUNT PLOT",size=25,color="red",family="cambria")
plt.show()
```

```python
sns.countplot(x=df["prediction"],edgecolor="k",saturation=0.5)
plt.title("COUNT PLOT",size=25,color="red",family="cambria")
plt.show()
```
