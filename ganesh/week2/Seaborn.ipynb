## hue

```python
sns.barplot(x="island",y="bill_depth_mm",data=data,hue="sex")
plt.show()
```

## order

```python
order_1=["Dream","Biscoe","Torgersen"]
sns.barplot(x="island",y="bill_depth_mm",data=data,hue="sex",order=order_1)
plt.show()
```

```python
order_1=["Dream","Biscoe","Torgersen"]
sns.barplot(x="island",y="bill_depth_mm",data=data,hue="sex",order=order_1,hue_order=["Female","Male"])
plt.show()
```

## confidence interval (ci)

```python
order_1=["Dream","Biscoe","Torgersen"]
sns.barplot(x="island",y="bill_depth_mm",data=data,hue="sex",order=order_1,hue_order=["Female","Male"],errorbar=("ci",100))
plt.show()
```

## n_boot

```python
order_1=["Dream","Biscoe","Torgersen"]
sns.barplot(x="island",
            y="bill_depth_mm",
            data=data,hue="sex",
            order=order_1,
            hue_order=["Female","Male"],errorbar=("ci",100),
            #errorbar=None,
            n_boot=7)
plt.show()
```

## orient="h" horizontal

```python
order_1=["Dream","Biscoe","Torgersen"]
sns.barplot(x="bill_length_mm",
            y="bill_depth_mm",
            data=data,orient="h")
plt.show()
```

## orient="v" vertical

```python
order_1=["Dream","Biscoe","Torgersen"]
sns.barplot(x="bill_length_mm",
            y="bill_depth_mm",
            data=data,orient="v")
plt.show()
```

```python
sns.barplot(x="island",y="bill_depth_mm",data=data,hue="sex",saturation=1)
plt.show()
```

## errcolor="color_name"

## capsize

```python
sns.barplot(x="island",y="bill_depth_mm",data=data,hue="sex",saturation=0.8,capsize=0.3)
plt.show()
```

# Strip Plot

A **strip plot** is very simple to understand. It is basically a scatter plot that differentiates different categories. Strip plots are considered a good alternative to a box plot or aviolin plot.

```python
df2=sns.load_dataset("tips")
```

```python
sns.set(style="whitegrid")
sns.stripplot(x="day",y="total_bill",data=df2,hue="sex",
              hue_order=["Female","Male"],palette="Reds",
              linewidth=0.5,edgecolor="w",size=7,marker="^") #jitter=1,alpha=0.5
plt.grid(True)
plt.show()
```

# Facet Grid  (Multiple Plots)

```python
df=sns.load_dataset("tips")
```

```python
df1=sns.FacetGrid(df,col="sex",hue="day")
df1.map(plt.scatter,"total_bill","tip").add_legend() 
plt.show()
```

```python
df1=sns.FacetGrid(df,col="day",hue="sex")
df1.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
df1=sns.FacetGrid(df,col="day",hue="sex")
df1.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
df1=sns.FacetGrid(df,col="sex",hue="day")
df1.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
df1=sns.FacetGrid(df,col="sex",hue="day",palette="plasma")
df1.map(plt.bar,"total_bill","tip",edgecolor="w").add_legend()
plt.show()
```
