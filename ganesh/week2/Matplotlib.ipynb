# MATPLOTLIB

#### Matplotlib is a popular Python library for creating static, animated, and interactive visualizations in Python. It provides a wide variety of plotting options, including line plots, scatter plots, bar charts, histograms, and more.

### Simple line plot

```python
from matplotlib import pyplot as plt
x=[0,1,2,3,4,5]
y=[0,10,20,30,40,50]
plt.plot(x,y,marker="*",mec="r",mfc="pink",ms=15)
#ms=marker size
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("THE LINE GRAPH")
plt.show()
print(x,y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,4,8,12,16])
print(x,y)
plt.plot(x,marker="*",mfc="black",ms="15")
plt.title("The Line Graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,5,2)
y=x**2    #x power 2
print(x,y)
plt.plot(x,y,marker="H",mec="black",mfc="yellow",ls="none",)
#mfc=marker face color
#mec=marker edge color
plt.title("The Line Graph")
plt.show()
```

# Subplots

```python
x=[1,2,3,4,5,6]
y1=[20,30,40,50,70,90]
y2=[10,30,40,60,80,100]

plt.suptitle("SUBPLOTS",c="r",size="16",family="cambria")  #for main title

plt.subplot(1,2,1)
plt.plot(x,y1,c="g",ls="dotted",marker="o",mfc="r")

plt.title("Plot-1",c="k",family="cambria")
plt.grid("both",ls="dashed",lw=2,c='k')

plt.subplot(1,2,2)
plt.plot(x,y2,c="red",ls="dashed",marker="o")

plt.title("Plot-2",c="k",family="cambria")
plt.grid("both",ls="none",lw=2,c='k')

plt.legend(["Percentage"],loc="upper center",framealpha=0.2,facecolor="pink",edgecolor="green",shadow=False,fancybox=False,)

plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.4, hspace=0.4)   #for space adjustment between the plots
plt.show()
```

# Stacked Bar Chart

##### A stacked bar chart is a type of bar chart that represents data in a stacked manner, where each bar is divided into segments or categories, and each segment represents a different subgroup or component of the whole. This type of chart is useful for showing the total size as well as the composition of each category.

```python
import matplotlib.pyplot as plt
```

## vertical stacked bar graph

```python
years=["2017","2018","2019","2020"]
placements=[150,275,324,295]
#dividing as branch wise placements
mech=[75,80,47,82]
it=[46,58,35,52]
ece=[67,59,72,62]

ece_start=[mech[i]+it[i] for i in range(len(mech))]
plt.title("PLACEMENTS COMPARISION USING STACKED BAR GRAPH",color="green")
plt.bar(years,mech,width=0.4,color="k",label="Mech")
plt.bar(years,it,bottom=mech,width=0.4,color="r",label="IT")
plt.bar(years,ece,bottom=ece_start,width=0.4,color="yellow",label="ECE")

plt.legend()

plt.ylim(0,250)   #y-axis limit
plt.show()
```

## Horizontal stacked bar graph

```python
years=["2017","2018","2019","2020"]
placements=[150,275,324,295]
#dividing as branch wise placements
mech=[75,80,47,82]
it=[46,58,35,52]
ece=[67,59,72,62]

plt.figure(figsize=(8,8))

ece_start=[mech[i]+it[i] for i in range(len(mech))]
plt.title("PLACEMENTS COMPARISION USING STACKED BAR GRAPH",color="green")
plt.barh(years,mech,height=0.4,color="k",label="Mech")
plt.barh(years,it,left=mech,height=0.4,color="r",label="IT")
plt.barh(years,ece,left=ece_start,height=0.4,color="yellow",label="ECE")

plt.legend()
plt.show()
```

# Grouped Bar Chart or Multiple Bar Plots

```python
import matplotlib.pyplot as plt
import numpy as np
years=["2017","2018","2019","2020"]
mech=[75,80,47,82]
it=[46,58,35,52]
ece=[67,59,72,62]
w=0.2
#note: if we not specify width by default it will be 0.8 

plt.figure(figsize=(10,7))   #Figure is the place where we are representing the plot

mech_bar=np.arange(len(years))
it_bar=[i+w for i in mech_bar]
ece_bar=[i+w for i in it_bar]

print(mech_bar)
plt.bar(mech_bar,mech,width=w,label="Mech",color="blue")
print(it_bar)
plt.bar(it_bar,it,width=w,label="IT",color="k")
print(ece_bar)
plt.bar(ece_bar,ece,width=w,label="ECE",color="yellow")

plt.xticks(mech_bar+w,years)

plt.ylim(0,100)
plt.title("Placement Comparison",color="k", family="cambria",size=25)
plt.ylabel("Placements",color="red",family="cambria",size=20)
plt.xlabel("Years",color="red",family="cambria",size=20)
plt.legend()
plt.show()

```

# Pie Chart

#### A pie chart is a circular statistical graphic that is divided into slices to illustrate numerical proportions. Each slice represents a proportionate part of the whole, and the size of each slice is proportional to the quantity it represents. Pie charts are commonly used to show the distribution of categories or percentages within a dataset. In Matplotlib, which is a popular Python plotting library, you can create a pie chart using the pie() function. The pie() function takes a set of values and automatically calculates the relative proportions, then creates a pie chart accordingly.

```python
import matplotlib.pyplot as plt

Student_Performance = ["Out Standing", "Excellent", "Good", "Average", "Poor"]
Student_Values = [50, 40, 30, 20, 10]

# Set the figure size
plt.figure(figsize=(8, 8))

# Create a pie chart
plt.pie(Student_Values, labels=Student_Performance, startangle=180, explode=[0.1, 0, 0, 0, 0], colors=["green", "orange", "yellow", "pink", "red"], shadow=False, autopct='%1.1f%%')

#autopct='%1.1f%%' in the plt.pie() function adds percentage labels to each slice.
#The format '1.1f' specifies one digit before the decimal point and one digit after.
#The startangle parameter in Matplotlib's pie() function is used to rotate the starting angle of the pie chart.
#The explode parameter in Matplotlib's pie() function is used to highlight or separate a particular slice from the rest of the pie chart.

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

# Histogram Chart

#### A histogram is a graphical representation of the distribution of a dataset. It displays the frequency of data points falling into specified ranges (bins). In Matplotlib, you can create a histogram using the hist() function.

```python
import matplotlib.pyplot as plt

Marks = [30, 45, 35, 60, 66, 70, 78, 80, 67, 90, 95, 100]
Grade_intervals = [0, 35, 50, 70, 100]

# Plot the histogram with specified histtype and color
plt.hist(Marks, Grade_intervals,histtype="barstacked", color="g", label="Marks Distribution")

plt.xticks([0, 35, 50, 70, 100])
plt.xlabel("Marks", size=15, color="r")
plt.ylabel("No. of Students", size=15, color="r")
plt.title("Class X", size=25)

# Add a legend
plt.legend()

plt.show()
```
