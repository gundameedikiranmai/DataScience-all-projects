**Object Detection:** 

Object detection is a technique that uses neural networks to localize and classifying objects in images. This computer vision task has a wide range of applications, from medical imaging to self-driving cars.

**Yolo:**

YOLO, which stands for "You Only Look Once," is a popular deep learning-based object detection algorithm. It's known for its speed and accuracy, making it widely used in various computer vision applications. The core idea behind YOLO is to perform object detection by dividing the input image into a grid and predicting bounding boxes and class probabilities for objects directly from the grid cells.

```python
from ultralytics import YOLO
import warnings
warnings.filterwarnings("ignore")
```

```python
model=YOLO("yolov8m.pt")
```

```python
result=model.predict(r"C:\Users\ganig\OneDrive\Desktop\Dataset\group_2.jpg",conf=0.5)
```

```python
result=result[0]
print(len(result.boxes))
```

```python
for box in result.boxes:
    label=result.names[box.cls[0].item()]
    cords=[round(x) for x in box.xyxy[0].tolist()]
    prob=box.conf[0].item()
    print("object type:",label)
    print("cooridinates:",cords)
    print("porbability:",prob)
    print("---")
```
