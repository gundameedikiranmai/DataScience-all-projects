## Splitting the Dataset into Training and Test Sets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_test.shape
```

```python
y_train.shape
```

## Feature Scaling

**Note:** The main purpose of **feature scaling** is to bring the feature values to definite range, so that computation happens faster.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test= scaler.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

## Training Dataset

```python
## Logistic Regression
from sklearn.linear_model import LogisticRegression
model_logistic = LogisticRegression() 
model_logistic.fit(x_train, y_train)
```

```python
y_pred_logistic=model_logistic.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm_logistic = confusion_matrix(y_test, y_pred_logistic)

print("Confusion Matrix (Logistic Regression):")
print(cm_logistic)
```

```python
plt.figure(figsize=(4,2))
sns.heatmap(cm_logistic, annot=True, cmap='viridis', fmt='g')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix Logistic Regression')
plt.show()
```

```python
## SVM(Support Vector Machine)
from sklearn.svm import SVC
model_svm = SVC()
model_svm.fit(x_train, y_train)
```

```python
y_pred_svm=model_svm.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm_svm = confusion_matrix(y_test, y_pred_svm)

print("Confusion Matrix (SVM):")
print(cm_svm)
```

```python
plt.figure(figsize=(4,2))
sns.heatmap(cm_svm, annot=True, cmap='plasma', fmt='g')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix SVM')
plt.show()
```

```python
## KNN (K-Nearest Neighbour)
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier()
model_knn.fit(x_train, y_train)
```

```python
y_pred_knn=model_knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm_knn = confusion_matrix(y_test, y_pred_knn)

print("Confusion Matrix (KNN):")
print(cm_knn)
```

```python
plt.figure(figsize=(4,2))
sns.heatmap(cm_knn, annot=True, cmap='Purples', fmt='g')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix KNN')
plt.show()
```

```python
## Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier()
model_tree.fit(x_train, y_train)
```

```python
y_pred_tree=model_tree.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm_tree = confusion_matrix(y_test, y_pred_tree)

print("Confusion Matrix (Decision Tree):")
print(cm_tree)
```

```python
plt.figure(figsize=(4,2))
sns.heatmap(cm_tree, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix Decision Tree')
plt.show()
```

```python
## Random Forest
from sklearn.ensemble import RandomForestClassifier
model_random=RandomForestClassifier()
model_random.fit(x_train,y_train)
```

## Ploting Accuracy of different ML Algorithms

```python
accuracy_values = [logistic_accuracy, svm_accuracy, knn_accuracy, decision_tree_accuracy, random_forest_accuracy, xgboost_accuracy]
labels = ["LR", "SVM", "KNN", "DTree", "RForest", "XGBoost"]

plt.figure(figsize=(10, 4))
bars = plt.bar(labels, accuracy_values,color="black")

# Add the values on top of the bars
for bar, value in zip(bars, accuracy_values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), round(value, 2), ha='center', va='bottom')

plt.title("Accuracy of different ML Algorithms")
plt.xlabel("Machine Learning Algorithm")
plt.ylabel("Accuracy")
plt.show()
```

**observations:**  
1. From the above bar graph we can summarize that SVM and Random Forest performed well with an accuracy of 88+ percentage
2. RForest > SVM > Logistic Regression > KNN > Decision Tree > XGBoost
